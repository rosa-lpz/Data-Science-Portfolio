## Load the data

### Subtask:
Load the data from the provided CSV file into a pandas DataFrame.


**Reasoning**:
Import pandas and load the data into a DataFrame, then display the first 5 rows.



```python
import pandas as pd

df = pd.read_csv('/pred_maint_timeseries_with_failures.csv')
display(df.head())
```

## Perform exploratory data analysis (eda)

### Subtask:
Perform exploratory data analysis (eda)


**Reasoning**:
Perform initial EDA steps by displaying column information, descriptive statistics, and checking for missing values.



```python
df.info()
display(df.describe())
display(df.isnull().sum())
```

**Reasoning**:
Visualize the distribution of the 'failure_warning' column to assess class imbalance.



```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 4))
sns.countplot(x='failure_warning', data=df)
plt.title('Distribution of Failure Warnings')
plt.xlabel('Failure Warning (0: No Failure, 1: Failure)')
plt.ylabel('Count')
plt.show()
```

**Reasoning**:
Create time series plots for temperature and pressure columns against the timestamp to observe trends. Convert 'datetimestamp' to datetime objects first.



```python
df['datetimestamp'] = pd.to_datetime(df['datetimestamp'])

temp_cols = [col for col in df.columns if 'enginegastemperature' in col]
fig, axes = plt.subplots(nrows=len(temp_cols) + 1, ncols=1, figsize=(12, 5 * (len(temp_cols) + 1)))

for i, col in enumerate(temp_cols):
    sns.lineplot(x='datetimestamp', y=col, data=df, ax=axes[i])
    axes[i].set_title(f'{col} over Time')
    axes[i].set_xlabel('Timestamp')
    axes[i].set_ylabel(col)

sns.lineplot(x='datetimestamp', y='pressure', data=df, ax=axes[-1])
axes[-1].set_title('Pressure over Time')
axes[-1].set_xlabel('Timestamp')
axes[-1].set_ylabel('Pressure')

plt.tight_layout()
plt.show()
```

**Reasoning**:
Generate a correlation heatmap to visualize the correlations between numerical features and the 'failure_warning' column.



```python
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

## Feature engineering

### Subtask:
Create new features from the existing time series data (e.g., lag features, rolling statistics, time-based features like hour of day, day of week, etc.).


**Reasoning**:
Extract time-based features, create lag features, and calculate rolling statistics from the relevant columns in the dataframe. Handle potential missing values resulting from lag and rolling calculations.



```python
# Ensure datetimestamp is datetime type
df['datetimestamp'] = pd.to_datetime(df['datetimestamp'])

# 1. Extract time-based features
df['hour_of_day'] = df['datetimestamp'].dt.hour
df['day_of_week'] = df['datetimestamp'].dt.dayofweek
df['day_of_year'] = df['datetimestamp'].dt.dayofyear

# Relevant sensor reading columns
sensor_cols = [col for col in df.columns if 'enginegastemperature' in col or col == 'pressure']

# 2. Create lag features
for col in sensor_cols:
    for lag in [1, 2, 3]:
        df[f'{col}_lag_{lag}'] = df.groupby('machineid')[col].shift(lag)

# 3. Calculate rolling statistics
window_size = 5
for col in sensor_cols:
    df[f'{col}_rolling_mean_{window_size}'] = df.groupby('machineid')[col].rolling(window=window_size).mean().reset_index(level=0, drop=True)
    df[f'{col}_rolling_std_{window_size}'] = df.groupby('machineid')[col].rolling(window=window_size).std().reset_index(level=0, drop=True)

# 4. (Skipped for now as EDA did not strongly suggest non-linear relationships requiring interaction/polynomial features)

# 5. Handle potential missing values introduced by lag or rolling features
# Filling with 0 is one simple approach, but forward fill or mean/median imputation might also be considered depending on the data characteristics.
df.fillna(0, inplace=True)

display(df.head())
display(df.tail())
```
